### Teoria Estymacji:
W praktyce, rozkład prawdopodobieństwa, nie jest znany.
Skończony zespół doświadczeń nazywamy próbą.
Wnioskowanie na podstawie próby o rozkładzie prawdopodobieństwa
(nieskończonego zespołu wszystkich możliwych doświadczeń) nazywamy
**estymacją**.

**Statystyka** tj. funckja zmiennych losowych z próbą.
Statystyka też jest zmienną losową.
 
**Estymator** T indexof n (x, .. xn, t) -> Statystyka o rozkładzie prawdopodobieństwa zależnym od pewnego parametru Theta.
Jaki powinien być dobry estymator?
Estymator nieobciążony - dla każdego n  E[ estymator T indexof n (Theta), który = Theta].
Obciążenie estymatora : B indexof n = E[T indexof n (Theta)] - Theta.
Estymator zgodny - estymator nieobiążony, którego wariancja dąży do 0 wraz z n.
Spełnia prawo wielkich liczb:
Istnieje takie e > 0  lim(inf) P(|Tn(Theta) - Theta| < e) = 1.
*Estymator to przepis na parametr Theta, który nas interesuje. Estymator najbardziej efektywny dla zbioru estymatorów tego samego parametru to ten, który ma najmniejszą wariancję*

### Przykłady:
Estymator wartości oczekiwanych:
X1, X2, X3, ..., X100

Tn(E(X)) === mean(X) = 1/x * SUM(X) spełnia prawo wielkich liczb, jest nieobciążony (jego obciążenie zbiega się do 0)

mean(x) ma rozkład normalny N(E(X), sigma(X) / sqrt(n))

--------------

Estymatory odchylenia standardowego pojedynczego pomiaru:

S(X) = sqrt(1/(n-1) * SUM_1..n(Xi - mean(X)**2) - zgodny, asymptotycznie nieobciążony.

S(X) = sqrt(1/n * SUM_1..n(Xi - mean(X)**2) - zgodny, asympt. nieobciążony, najefektywniejszy

S(x) = k indexof n * S(x)   k indexof n = sqrt((n-1) / 2) * poprawka((n-1)/2) / poprawka(n/2)
